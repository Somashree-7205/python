## Outlier Treatment 

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lr_age,ur_age=remove_outlier(df['referral_age'])
df['referral_age']=np.where(df['referral_age']>ur_age,ur_age,df['referral_age'])
df['referral_age']=np.where(df['referral_age']<lr_age,lr_age,df['referral_age'])

lr_exp_in_yrs,ur_exp_in_yrs=remove_outlier(df['referral_exp_in_years'])
df['referral_exp_in_years']=np.where(df['referral_exp_in_years']>ur_exp_in_yrs,ur_exp_in_yrs,df['referral_exp_in_years'])
df['referral_exp_in_years']=np.where(df['referral_exp_in_years']<lr_exp_in_yrs,lr_exp_in_yrs,df['referral_exp_in_years'])

lr_curr_sal,ur_curr_sal=remove_outlier(df['referral_current_salary'])
df['referral_current_salary']=np.where(df['referral_current_salary']>ur_curr_sal,ur_curr_sal,df['referral_current_salary'])
df['referral_current_salary']=np.where(df['referral_current_salary']<lr_curr_sal,lr_curr_sal,df['referral_current_salary'])



lr_10th,ur_10th=remove_outlier(df['referral_10_th_marks'])
df['referral_10_th_marks']=np.where(df['referral_10_th_marks']>ur_10th,ur_10th,df['referral_10_th_marks'])
df['referral_10_th_marks']=np.where(df['referral_10_th_marks']<lr_10th,lr_10th,df['referral_10_th_marks'])


lr_12th,ur_12th=remove_outlier(df['referral_12th_marks'])
df['referral_12th_marks']=np.where(df['referral_12th_marks']>ur_12th,ur_12th,df['referral_12th_marks'])
df['referral_12th_marks']=np.where(df['referral_12th_marks']<lr_12th,lr_12th,df['referral_12th_marks'])



lr_grd_mark,ur_grd_mark=remove_outlier(df['referral_graduation_marks'])
df['referral_graduation_marks']=np.where(df['referral_graduation_marks']>ur_grd_mark,ur_grd_mark,df['referral_graduation_marks'])
df['referral_graduation_marks']=np.where(df['referral_graduation_marks']<lr_grd_mark,lr_grd_mark,df['referral_graduation_marks'])

lr_exp_sal,ur_exp_sal=remove_outlier(df['referral_expected_sal'])
df['referral_expected_sal']=np.where(df['referral_expected_sal']>ur_exp_sal,ur_exp_sal,df['referral_expected_sal'])
df['referral_expected_sal']=np.where(df['referral_expected_sal']<lr_exp_sal,lr_exp_sal,df['referral_expected_sal'])



lr_hr_score,ur_hr_score=remove_outlier(df['referral_HR_Interview_score'])
df['referral_HR_Interview_score']=np.where(df['referral_HR_Interview_score']>ur_hr_score,ur_hr_score,df['referral_HR_Interview_score'])
df['referral_HR_Interview_score']=np.where(df['referral_HR_Interview_score']<lr_hr_score,lr_hr_score,df['referral_HR_Interview_score'])


lr_sal_ofr,ur_sal_ofr=remove_outlier(df['Salary offered'])
df['Salary offered']=np.where(df['Salary offered']>ur_sal_ofr,ur_sal_ofr,df['Salary offered'])
df['Salary offered']=np.where(df['Salary offered']<lr_sal_ofr,lr_sal_ofr,df['Salary offered'])



```
