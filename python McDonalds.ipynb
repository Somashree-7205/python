### 2) find out the items which have less than  1 of total fat ???

```python
df1=df[df['Total Fat']>1].sort_values(by='Total Fat',ascending=False).head(10)
```

```python
plt.figure(figsize=(12,4))
sns.boxplot(x='Total Fat',y='Category',data=df1,color='pink');
```

### 3)  find out the top  5 breakfast items which have more protein??

```python
df1=df[df['Category']=='Breakfast']
```

```python
df1.sort_values(by='Protein',ascending=False,inplace=True)
```

```python
plt.figure(figsize=(10,3))

sns.barplot(x='Item',y='Protein',data=df1,ci=None);

plt.xticks(rotation=90);
```
